```python
plt.pie(data["Sex"].value_counts(),explode=[0,0.05],autopct="%1.2f%%",labels=["Female","Male"])
plt.title("sex of people")
plt.show()
```

```python
sns.countplot(x='Sex',data=data,hue="Sex",palette='magma',legend=False,color='r')
plt.xlabel("gender")
plt.ylabel("number of people")
plt.show()
```

```python
sns.countplot(x='Pclass',data=data,hue='Pclass',palette='magma')
plt.xlabel("Pclass")
plt.ylabel("Number of pepole")
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5))
sns.countplot(x='SibSp',data=data,hue='SibSp',ax=axes[0],palette='plasma')
sns.countplot(x='Parch',data=data,hue='Parch',ax=axes[1],palette='mako')
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,7))
sns.histplot(data['Fare'],kde=True,ax=axes[0])
sns.histplot(data['Age'].dropna(),kde=True,ax=axes[1])
plt.show()
```
